```python
# Earn optimal returns at a targeted risk level!
# This script shows a return of stock share in a portfolio!
import yfinance as yf
import riskfolio as rp
import warnings
warnings.filterwarnings("ignore")

assets = ["ETH-USD", "COIN", "AAPL", "TSLA", "FCX"]
assets.sort()
data = yf.download(assets, start ="2022-01-01", end = "2022-9-30")
returns = data['Adj Close'].pct_change().dropna()

port = rp.Portfolio(returns = returns)

port.assets_stats(method_mu = 'hist', method_cov = 'hist', d = 0.94)

w_rp = port.rp_optimization(model = "Classic", rm = "MV", hist = True,
rf = 0, b = None)

ax = rp.plot_pie(w = w_rp)
```
